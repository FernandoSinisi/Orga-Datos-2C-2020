```python
import pandas as pd
import datetime as dt

data = {
        'NombreColumna1': [7,20,15,22],
        'NombreColumna2': ['PID_1','PID_2','PID_3','PID_9']
}
data_2 = {
        'Value2': [17,15,10,1,3, 11, 13, 20],
        'ProductId': ['PID_1','PID_2','PID_3','PID_4','PID_5', 'PID_6', 'PID_7', 'PID_8']
}
df = pd.DataFrame(data, columns = data.keys())
df2 = pd.DataFrame(data_2, columns = data_2.keys())
df
```

```python
# Reordenar columnas
nuevo_orden = ['NombreColumna2','NombreColumna1']
df = df[nuevo_orden]
df
```

```python
# Renombrar columnas
df.columns = ['ProductId','value']
df
```

```python
#Merge izquierda
df_merge_left = df.merge(df2,how='left', on = 'ProductId')
df_merge_left
```

```python
#Merge derecha
df_merge_right = df.merge(df2,how='right', on = 'ProductId')
df_merge_right
```

```python
#Merge interseccion
df_merge_inner = df.merge(df2, on = 'ProductId')
df_merge_inner
```

```python
#Merge todo
df_merge_outer = df.merge(df2,how = 'outer', on = 'ProductId')
df_merge_outer
```

```python
# stack, llevar columnas a indice en la fila (interno)
df_stack = df.stack()
df_stack
```

```python
#Unstack, llevar indice de filas a columna, inversa de stack
df_unstack_1 = df_stack.unstack()
df_unstack_1
```

```python
#Unstack, llevar indice de filas a columna, inversa de stack (ejecutado 2 veces sobre un stack)
df_unstack_2 = df_unstack_1.unstack()
df_unstack_2
```
