Creacion de un modelo de Random forest para seleccionar los features mas relevantes de los encontrados

```python
RF_clf = RandomForestClassifier(n_estimators=10000,random_state=42,n_jobs=-1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_train, train.Target, test_size=0.33, random_state=7)
```

```python
RF_clf.fit(x_train, y_train)
```

```python
thresholds = sort(RF_clf.feature_importances_)
for thresh in thresholds:
    selection = SelectFromModel(RF_clf, threshold=thresh, prefit=True)
    select_X_train = selection.transform(x_train)
 
    selection_model = RandomForestClassifier(n_estimators=10000,random_state=42,n_jobs=-1)
    selection_model.fit(select_X_train, y_train)
    
    select_X_test = selection.transform(x_test)
    y_pred = selection_model.predict(select_X_test)
    predictions = [round(value) for value in y_pred]
    accuracy = accuracy_score(y_test, predictions)
    logloss = log_loss(y_test, predictions)
    print("Thresh=%.3f, n=%d, Accuracy: %.2f%%, logloss: %.2f" % (thresh, select_X_train.shape[1], accuracy*100.0,logloss))
```
